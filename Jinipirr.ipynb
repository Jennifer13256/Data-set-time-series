```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Jennifer13256/Data-set-time-series/main/Electric_Production.csv.csv")
```

```python
df.head()
```

```python
date_rng = pd.date_range(start='01-01-1985',end='05-01-1985',freq='D')
```

```python
time_series_data = pd.Series(range(0,11))
```

```python
df
```

```python
print("Temporal Order:")
```

```python
print(time_series_data)
```

```python
import matplotlib.pyplot as plt
```

```python
time_series_data_trend = time_series_data + 0.5 * pd.Series(range(0,11))
```

```python
df
```

```python
plt.plot(time_series_data_trend)
```

```python
plt.title("Time Series Data with Trend")
```

```python
plt.show()
```

```python
time_series_data_seasonal = time_series_data +2 * pd.Series([1,-1]*5, index=date_rng)
```

```python
plt.plot(time_series_data_seasonal)
```
